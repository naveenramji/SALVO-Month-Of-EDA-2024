```python
#importing necessary modules
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the input file
df = pd.read_csv('../input/new-york-dataset/new_york_listings_2024.csv')
print(df)
```

```python
#deleting null and duplicate values
df.drop_duplicates(inplace=True)
df.dropna(inplace = True)
```

```python
#frequency chart for room types
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='room_type')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
#price distribution of rooms
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='bedrooms', y='price')
plt.title('Bedrooms vs. Price')
plt.xlabel('Bedrooms')
plt.ylabel('Price ($)')
plt.show()
```

```python
#plot based on location of each room
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type', palette='viridis', alpha=0.7)
plt.title('Airbnb Listings in New York City')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Room Type')
plt.show()
```
